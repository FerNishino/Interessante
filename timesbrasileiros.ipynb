```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/brasileirao-2004-2022.csv")
df
```

```python
df['win_percentage'] = (df[' won'] / df[' played']) * 100

# Agrupar os dados por temporada e calcular a média da frequência de vitória
season_win_avg = df.groupby('season')['win_percentage'].mean()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(season_win_avg.index, season_win_avg.values, color='blue')
plt.title('Frequência Média de Vitória por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Frequência Média de Vitória (%)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
# Encontrar os clubes com aproveitamento de vitória em todas as temporadas
clubs_with_all_wins = df.groupby(' team')['win_percentage'].count() == df['season'].nunique()
clubs_with_all_wins = clubs_with_all_wins[clubs_with_all_wins].index.tolist()

# Filtrar o DataFrame para incluir apenas os clubes com aproveitamento de vitória em todas as temporadas
df_filtered = df[df[' team'].isin(clubs_with_all_wins)]

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
for club in df_filtered[' team'].unique():
    club_data = df_filtered[df_filtered[' team'] == club]
    plt.plot(club_data['season'], club_data['win_percentage'], label=club)

plt.title('Aproveitamento de Vitória ao Longo das Temporadas')
plt.xlabel('Temporada')
plt.ylabel('Frequência de Vitória (%)')
plt.legend(loc='upper left')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
# Encontrar os times que marcaram mais gols
top_scorers = df.groupby(' team')[' goals_for'].sum().nlargest(10)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
top_scorers.plot(kind='bar', color='green')
plt.title('Times que Marcaram Mais Gols')
plt.xlabel('Time')
plt.ylabel('Total de Gols Marcados')
plt.xticks(rotation=45)
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
# Encontrar os times que sofreram mais gols
top_conceders = df.groupby(' team')[' goals_against'].sum().nlargest(10)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
top_conceders.plot(kind='bar', color='red')
plt.title('Times que Sofreram Mais Gols')
plt.xlabel('Time')
plt.ylabel('Total de Gols Sofridos')
plt.xticks(rotation=45)
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
